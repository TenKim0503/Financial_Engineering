```python
import numpy as np #import numpy
import random as ra #import random
from scipy.stats import norm #Normal Distribution from scipy.stats
import itertools
import matplotlib.pyplot as plt
```

```python
def random_List(size):
  result=[]

  for v in range(size):
    result.append(np.round(ra.uniform(0.1,1.5),2))

  return result
```

```python
def movement(M,N,s0,r,sig,K,T):
  S=np.random.normal(size=(M,N))

  for i in range(M):
    for j in range(N):
      if j == 0:
        S[i,j] = s0 *(1+ r*T/N+ sig * np.sqrt(T/N)*S[i,j])
      else:
        S[i,j] = S[i,j-1] * (1+ r*T/N +sig* np.sqrt(T/N)*S[i,j]) 
  for i in range(M):
   plt.plot([s0]+list(S[i,:]))
   plt.xlabel('TIME')
   plt.ylabel('PRICE')
```

```python
r=random_List(3)
sig=random_List(3)
```

```python
result = list(itertools.product(*[r,sig]))
```

```python
for x,y in result:
  for z in range(9):
    plt.figure(z)
    plt.text(0,200,'(Risk,Volaltility)'+':'+str(result[z]))
    plt.text(0,220,'S0=100'+','+'K=100'+','+'T=30/365')
    plt.title('THE MOVEMENT OF PRICE'+'-'+str(z))
    movement(100,30,100,x,y,100,30/365)
    
```
